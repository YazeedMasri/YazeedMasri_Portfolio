```python
#Creating a function that returns matrix based on day as a np array
def conv_to_npa(day_list):
    
    try:
        # Filters df27 values based on list values
        df00 = df27[day_list]
        # Transposes to filter rows as well
        df0x = df00.T[day_list]
    except:
        print('Make sure the input is a list with values in df22')
    
    return np.array(df0x)
```

# Creating the distance functions - for two points and for a list of points

```python
#Function to calculate distance between two locations/OBID
def distance(OBID_1, OBID_2):
    
    # obtains location of the OBIDs as an index
        ind1 = df22.columns.get_loc(OBID_1)
        ind2 = df22.columns.get_loc(OBID_2)-1
        
    # returting the distance as defined in np array 1
        return n1[ind2, ind1]
    

#Function to calculate total distance of a list
def Tot_dis(x):
    total_distance = 0
    
    for i in range(len(x) - 1):
        

        # get the object IDs for the current pair of objects
            obj1 = x[i]
            obj2 = x[i+1]
            
        # calculate the distance between the objects using the distance function
            dist = distance(obj1, obj2)
            
        # add the distance to the total distance
            total_distance += dist
            

        
    # add the distance between the last and first objects in the permutation
    obj1 = x[-1]
    obj2 = x[0]

    return round(total_distance, 2)
```

# Using the TSP module

```python
#Creating an optimal path function that returns optimal path and total distance
def Opt_path(distance, locations, Day):
# solving to find the most optimal permutation
    path = solve_tsp(distance)
# initializing a final list to return 
    F_list = []
    
# looping through values to find the locations and add to the list
    for i in range(len(path)):
        F_list.append(locations[path[i]])
# finding total distance using the total distance function and setting output
    print(f' For Day {Day}, \n The optimal path is: {F_list} \n \nTotal distance: {Tot_dis(F_list)} minutes \n')
    
    
Opt_path(npD1, Day_1, 1)
Opt_path(npD2, Day_2, 2)
Opt_path(npD3, Day_3, 3)
Opt_path(npD4, Day_4, 4)
Opt_path(npD5, Day_5, 5)
```

```python
#Same function that returns path only
def O_path(distance, locations):
# solving to find the most optimal permutation
    path = solve_tsp(distance)
# initializing a final list to return 
    F_list = []
    
# looping through values to find the locations and add to the list
    for i in range(len(path)):
        F_list.append(locations[path[i]])
    return F_list

OP_D1 = O_path(npD1, Day_1)
OP_D2 = O_path(npD2, Day_2)
OP_D3 = O_path(npD3, Day_3)
OP_D4 = O_path(npD4, Day_4)
OP_D5 = O_path(npD5, Day_5)
```

```python
Cust_sheet = pd.read_excel('/Users/yazeedmasri/downloads/customer_list.xlsx', sheet_name="Customers", index_col = None)
df09 = pd.DataFrame(data = Cust_sheet)
```

```python
df08 = df09[['OBJECTID','Latitude','Longitude']]
np08 = np.array(df08)
```

```python
#Creating coordinates dictionary to store OBID and corresponding coordinated
coordinates = {}
for i in np08:
    coordinates[int(i[0])] = (i[1], i[2])

#Function creates a list of tuples that correspond to optimal paths 
def OP_coords(lst):
    D_coord = []
    for i in lst:
        D_coord.append(coordinates[i])
    return D_coord

D1_coords = OP_coords(OP_D1)
D2_coords = OP_coords(OP_D2)
D3_coords = OP_coords(OP_D3)
D4_coords = OP_coords(OP_D4)
D5_coords = OP_coords(OP_D5)
```

```python
#Calculating the center coordinates
center = np.mean(D1_coords, axis=0)

#Creating folium map object
m = fl.Map(location=center, zoom_start=10)

# Displaying the line to visualize the path
fl.PolyLine(locations=D1_coords, color='red').add_to(m)


# Marking first and last locations
fl.Marker(location=[D1_coords[0][0], D1_coords[0][1]], icon=fl.Icon(color='green')).add_to(m)
fl.Marker(location=[D1_coords[-1][0], D1_coords[-1][1]], icon=fl.Icon(color='red')).add_to(m)


#Looping through Day coordinates to map
for coord in D1_coords[0:-1]:
    fl.Marker(location=[coord[0], coord[1]]).add_to(m)
    
#Displaying map
m
```
